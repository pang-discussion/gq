```python
# 导包
import torch
import numpy as np
import torch.nn as nn
import scipy
import random
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
import scipy.spatial

np.random.seed(0)
```

```python
embeddings = np.load("word2vec_wikitext_embedding.npy")

word_to_idx = np.load("word2vec_wikitext_vocab.npy",allow_pickle=True).item()

idx_to_word = {value: key for key, value in word_to_idx.items()}
```

```python
def get_idx(word):
    idx = 0
    if word in word_to_idx.keys():
        idx = word_to_idx[word]
    else:
        idx = word_to_idx['<unk>']
    return idx

def get_idxes(words):
    idx = []
    for i in words:
        if i in word_to_idx.keys():
            idx.append(word_to_idx[i])
        else:
            idx.append(word_to_idx['<unk>'])
    return idx

def find_nearest(word, embedding_weights):
    index = get_idx(word)
    embedding = embedding_weights[index]
    cos_dis = np.array([scipy.spatial.distance.cosine(e, embedding) for e in embedding_weights])
    return [idx_to_word[i] for i in cos_dis.argsort()[:10]]
```

```python
words = ['red', 'green', 'computer']
idx = get_idxes(words)
vectors = []
for i in idx:
    vectors.append(embeddings[i])
input1 = torch.tensor(vectors[0]).resize(1, 300)
input2 = torch.tensor(vectors[1]).resize(1, 300)
input3 = torch.tensor(vectors[2]).resize(1, 300)
cos = nn.CosineSimilarity(dim=1)
print(cos(input1, input2))
print(cos(input1, input3))
print(cos(input3, input2))
```

```python
for word in ["two", "green", "computer"]:
    print(word, find_nearest(word, embedding_weights=embeddings))
```

```python
indexs = ['one', 'eight', 'three', 'two', 'four', 'seven', 'six', 'nine', 'five',
         'green', 'blue', 'white', 'red', 'black', 'yellow', 'brown', 'grey',
         'computer', 'entertainment', 'program', 'engineering', 'gameplay','nintendo', 'technology']
random.shuffle(indexs)
ids = get_idxes(indexs)

wordvectors = []
for i in ids:
    wordvectors.append(embeddings[i])
wordvectors = np.array(wordvectors)
```

```python
pca = PCA(n_components=2)
wordvectors = pca.fit_transform(wordvectors)
wordvectors = pd.DataFrame(wordvectors)
from sklearn.cluster import KMeans
classCount = 3
clf = KMeans(n_clusters=classCount)
clf.fit(wordvectors)
predicts = clf.predict(wordvectors)
for i, j in zip(indexs, clf.labels_):
    print(i, ":", j, end='  ')
print()

plt.figure(dpi=120, figsize=(6, 6))
plt.scatter(x=wordvectors[0],y=wordvectors[1],s=10, c=predicts)

for i, label in enumerate(indexs):
    plt.annotate(label, (wordvectors[0][i],wordvectors[1][i]), fontsize=8)
plt.show()
```
